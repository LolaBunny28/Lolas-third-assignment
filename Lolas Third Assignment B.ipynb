```python
import sklearn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import datetime
```

```python
CarsPrice = pd.read_csv('Automobile Price Data.csv')
CarsPrice.head(5)
```

```python
CarsPrice.shape
```

```python
Replace the '?' value with NaN values
```

```python
CarsPrice = CarsPrice.replace('?', np.nan)
CarsPrice.head(5)
```

```python
CarsPrice = CarsPrice.dropna()
```

```python
CarsPrice.head(5)
```

```python
CarsPrice.isna().sum()
```

```python
CarsPrice.dtypes
```

```python
CarsPrice1 = CarsPrice[CarsPrice['normalized-losses']!='?']
normalised_mean = CarsPrice['normalized-losses'].astype(int).mean()
CarsPrice['normalized-losses'] = CarsPrice['normalized-losses'].replace('?',normalised_mean).astype(int)
```

```python
CarsPrice1 = CarsPrice[CarsPrice['price']!='?']
normalised_mean = CarsPrice1['price'].astype(int).mean()
CarsPrice['price'] = CarsPrice['price'].replace('?',normalised_mean).astype(int)
```

```python
CarsPrice1 = CarsPrice[CarsPrice['peak-rpm']!='?']
normalised_mean = CarsPrice1['peak-rpm'].astype(int).mean()
CarsPrice['peak-rpm'] = CarsPrice['peak-rpm'].replace('?',normalised_mean).astype(int)
```

```python
CarsPrice1 = CarsPrice[CarsPrice['horsepower']!='?']
normalised_mean = CarsPrice1['horsepower'].astype(int).mean()
CarsPrice['horsepower'] = CarsPrice['horsepower'].replace('?',normalised_mean).astype(int)
```

```python
CarsPrice1 = CarsPrice[CarsPrice['bore']!='?']
normalised_mean = CarsPrice1['bore'].astype(float).mean()
CarsPrice['bore'] = CarsPrice['bore'].replace('?',normalised_mean).astype(float)
```

```python
CarsPrice1 = CarsPrice[CarsPrice['stroke']!='?']
normalised_mean = CarsPrice1['stroke'].astype(float).mean()
CarsPrice['stroke'] = CarsPrice['stroke'].replace('?',normalised_mean).astype(float)
```

```python
CarsPrice['num-of-doors'] = CarsPrice['num-of-doors'].replace('?','four')
CarsPrice.head()
```

```python
CarsPrice.drop(['bore', 'stroke', 'compression-ratio'], inplace=True, axis=1)
CarsPrice.head()
```

```python
CarsPrice.drop(['make', 'aspiration', 'drive-wheels', 'body-style'], inplace=True, axis=1)
CarsPrice.head()
```

```python
CarsPrice.drop(['symboling', 'fuel-type', 'fuel-system'], inplace=True, axis=1)
CarsPrice.head()
```

```python
CarsPrice.describe()
```

```python
CarsPrice.to_csv('CarsPriceCleaned.csv', index=False)
```
